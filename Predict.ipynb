```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dia_data = pd.read_csv('heart.csv')
dia_data
```

```python
dia_data.loc[dia_data["Sex"] == "M", "Sex"] = 1
dia_data.loc[dia_data["Sex"] == "F", "Sex"] = 0
dia_data
```

```python
dia_data.loc[dia_data["ST_Slope"] == "Up", "ST_Slope"] = 1
dia_data.loc[dia_data["ST_Slope"] == "Flat", "ST_Slope"] = 0
dia_data.loc[dia_data["ST_Slope"] == "Down", "ST_Slope"] = -1
dia_data
```

```python
dia_data.loc[dia_data["ExerciseAngina"] == "Y", "ExerciseAngina"] = 1
dia_data.loc[dia_data["ExerciseAngina"] == "N", "ExerciseAngina"] = 0
dia_data
```

```python
dia_data.loc[dia_data["RestingECG"] == "Normal", "RestingECG"] = 0
dia_data.loc[dia_data["RestingECG"] == "ST", "RestingECG"] = 1
dia_data.loc[dia_data["RestingECG"] == "LVH", "RestingECG"] = 2
dia_data
```

```python
dia_data.loc[dia_data["ChestPainType"] == "ATA", "ChestPainType"] = 0
dia_data.loc[dia_data["ChestPainType"] == "NAP", "ChestPainType"] = 1
dia_data.loc[dia_data["ChestPainType"] == "ASY", "ChestPainType"] = 2
dia_data.loc[dia_data["ChestPainType"] == "TA", "ChestPainType"] = 3
dia_data
```

```python
X = dia_data.drop(columns=['HeartDisease'])
X
```

```python
y = dia_data['HeartDisease']
y
```

```python
model = DecisionTreeClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
```

```python
score = accuracy_score(y_test,predict)
score
```
